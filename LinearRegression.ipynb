```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Simple linear regression.csv")
```

```python
print(data)
```

```python
plt.scatter(data.SAT, data.GPA)
```

```python
def Gradient(w, b, data):
    daba_w = 0
    daba_b = 0
    m = len(data) - 1
    for i in range(m-1):
        y = data.iloc[i].GPA
        x = data.iloc[i].SAT

        daba_w += (w*x + b - y)*x
        daba_b += (w*x + b - y)
    
    daba_w = daba_w / m
    daba_b = daba_b / m

    return daba_w, daba_b
```

```python
w = 0
b = 0
epochs = 1000
alpha = 0.001

for iter in range(epochs):
    daba_w, daba_b = Gradient(w, b, data)
    print(w)
    w -= alpha*daba_w
    b -= alpha*daba_b
```
